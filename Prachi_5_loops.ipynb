### 41. Given a dictionary of 10 items, delete all items whose value is less than 50 using a for loop

```python
marks = {"a":80, "b":78, "c":49, "d":45, "e":67, "f":59, "g":89, "h":77, "i":90, "j":43}

keys_to_remove = []

for key, value in marks.items():
    if value > 50:
        keys_to_remove.append(key)

for key in keys_to_remove:
    del marks[key]

print(marks)
```
