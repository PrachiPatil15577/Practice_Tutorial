### 36. Determine the type of triangle using sides: acute, right, obtuse.

```python
a = int(input("enter a side a"))
b = int(input("enter a side b"))
c = int(input("enter a side c"))

sides = sorted([a,b,c])
x, y, z = sides

if x + y <= z:
    print("not valid triangle")
elif z**2 == x**2 + y**2:
    print("Obtuse angle")
elif z**2 > x**2 + y**2:
    print("Right angle")
else:
    print("Acute angle")
```
