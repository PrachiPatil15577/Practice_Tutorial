## **2. Find the sum of all numbers inside a nested list.**
Example: `[[10,20], [30], [5, 7, 8]]`

````python
# Write your answer here

lst = [[10,20],[30],[5,7,8]]
sum = 0

for num in lst:
    for item in num:
        #print(item)
        sum += item

print("sum is :", sum)
````

## **3. Count how many even numbers are present in a nested list.**

````python
# Write your answer here
lst = [[10,20],[30],[5,7,8]]
count = 0

for i in lst:
    for n in i:
        if n % 2 == 0:
         count += 1

print("even num:", count)
````

## **4. Convert all strings to uppercase inside a nested list.**
Example: `[['hello', 'world'], ['python']]`

````python
# Write your answer here
ch = [['hello', 'world'], ['python']]

for n in ch:
    for i in range(len(n)):
        n[i] = n[i].upper()

print(ch)
````

## **7. Find the largest number inside a nested list.**

````python
# Write your answer here

lst = [[10,20], [50,80], [75,45]]

largest = 0


for inner_lst in lst:
    for val in inner_lst:
        if val > largest:
            largest = val

print(largest)


````

## **8. Flatten a nested list into a single list.**
`[[1,2], [3,4], [5]] â†’ [1,2,3,4,5]`

````python
# Write your answer here

lst = [[1,2], [3,4], [5]]

flat_lst = []

for ch in lst:
    for item in ch:
        flat_lst.append(item)

print(flat_lst)
````

## **9. Replace all negative numbers with 0 inside a nested list.**

````python
# Write your answer here

lst = [[1,2,-1], [-2,5], [4,7,-2]]

for a in range(len(lst)):
    for b in range(len(lst[a])):
        if lst[a][b] < 0:
            lst[a][b] = 0

            
print(lst)
            
````

## **19. Find total stock by summing all 'quantity' values in a nested dictionary.**

````python
# Write your answer here

stock = {
    "prod1": {"name":"pen", "quantity":100},
    "prod2": {"name":"pencil", "quantity":50},
    "prod3": {"name":"sharpner", "quantity":10}
}

total = 0

for val in stock.values():
    total += val["quantity"]

print("total is:", total)
````

## **20. Convert all inner dictionary keys to uppercase.**

````python
# Write your answer here

stock = {
    "prod1": {"name":"pen", "quantity":100},
    "prod2": {"name":"pencil", "quantity":50},
    "prod3": {"name":"sharpner", "quantity":10}
}

new_dt = {}

for outer_key, details in stock.items():
    inner_key = {}
    for  key,value in details.items():
        inner_key[key.upper()] = value
        #print(inner_key)

    new_dt[outer_key] = inner_key

new_dt
````
