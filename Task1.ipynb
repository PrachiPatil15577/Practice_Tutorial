```python
# Join transaction with txn code lookup
tran_with_desc = acqr_tran_details_transaction.merge(
    sastpit_txn_code,
    how="left",
    left_on="transaction_type",
    right_on="txn_code"
)
```

```python
tran_with_desc.head()
```

```python
#Add name of person from PAPITDTLSD0 using PAN
tran_with_person = tran_with_desc.merge(
    papitdtlsd0,
    how="left",
    left_on="pan",
    right_on="pan",
    suffixes=("", "_person")
)

tran_with_person.head()
```

```python
# Join with holdings on PAN + account + ISIN
tran_with_holdings = tran_with_person.merge(
    acqr_tran_hldng_soh_sot_detail,
    how="left",
    left_on=["pan", "beneficiary_account_number", "isin"],
    right_on=["pan_acquirer", "beneficiary_account_number", "isin"],
    suffixes=("", "_hold")
)
```

```python
# Join with ISIN capital details
final_df = tran_with_holdings.merge(
    hst_isincapdtlsd0,
    how="left",
    left_on="isin",
    right_on="isin",
    suffixes=("", "_capital")
)

final_df.head()
```

```python
final_df["computed_percent_equity"] = (
    final_df["closing_bal_on_transact_day"] / final_df["issued_capital"] * 100
)

final_df[[
    "pan", "beneficiary_account_number", "isin",
    "closing_bal_on_transact_day", "issued_capital",
    "percnt_tot_equity_share", "computed_percent_equity"
]].head()
```
