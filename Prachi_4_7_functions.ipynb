### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
def sum(a, b):
    return a+b 

def sub(a, b):
    return a-b 

def mul(a, b):
    return a*b

def div(a, b):
    return a/b

def adapter_func(func, a, b):
    return func(a, b)

result1 = adapter_func(sum, 10, 20)
result2 = adapter_func(sub, 15, 5)
result3 = adapter_func(mul, 4, 5)
result4 = adapter_func(div, 10, 2)

print(result1)
print(result2)
print(result3)
print(result4)
```

### Write a function process_students(data) that prints total marks and grade for each student.

- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

### Create a function apply_discounts(products) that updates price after discount. 
- Example: [ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ] 
- Expected Output: Laptop final: 45000 Phone final: 19000

```python
def apply_disc(products):
    for data in products:
        name = data["name"]
        price = data["price"]
        discount = data["discount"]
        
        update_price = price - (price * discount / 100)
        print(f"product_name : {name}, Original_val : {price}, updated_val : {update_price}")


product_data = [
    { "name":"Laptop", "price": 50000, "discount":10},
    { "name":"Phone", "price": 20000, "discount":5},
]

apply_disc(product_data)

```

### Write out_of_stock(items) that returns all products with stock < 20.

```python
def out_of_stock(items):
    low_stock_items = []

    for item in items:
        if item['stock'] < 20:
            low_stock_items.append(item)

    return low_stock_items
products = [
    {"name": "mouse", "stock": 20},
    {"name": "printer", "stock":50},
    {"name": "keyboard", "stock": 15}
]

out_of_stock(products)
```

### Write search_product(items, name) → return matching product(s).

```python

def search_product(item,name):
    for product in item:
        if product['name'] == name:
            return product
    return "Product not found"
item_list = [
    {'name': 'laptop', 'price': 50000},
    {'name': 'Phone', 'price': 20000},
    {'name': 'Tablet', 'price': 15000}]
search_product(item_list, 'Phone')
```
